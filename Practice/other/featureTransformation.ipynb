# Transforming Nominal Attributes

```python
dataSetNo.columns
```

```python
mJob_field=dataSetNo.Mjob
print(mJob_field)
```

```python
mJob_field_unique=mJob_field.unique()
print(mJob_field_unique)
```

```python
from sklearn.preprocessing import LabelEncoder
gle = LabelEncoder()
mJob_labels = gle.fit_transform(mJob_field)
mJob_mappings = {index: label for index, label in enumerate(gle.classes_)}
mJob_mappings
```

```python
dataSetNo['Mjob'] = mJob_labels
dataSetNo.head()
```
