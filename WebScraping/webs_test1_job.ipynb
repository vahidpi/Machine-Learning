```python
import requests # to send a request to a web site and recive html
```

```python
from bs4 import BeautifulSoup  # https://www.crummy.com/software/BeautifulSoup/bs4/doc/
```

```python
page = requests.get("https://www.monster.com/jobs/search?q=software+engineer&where=Seattle")
# page_st = requests.get("https://www.technojobs.co.uk/search.phtml/software-engineer/searchfield/locationfinland/radius25/salary0")
```

```python
# page
# page_st
```

```python
soup=BeautifulSoup(page.text,"html.parser")
# soup_st = BeautifulSoup(page_st.text, "html.parser")
```

```python
soup.title
```

```python
soup.div
```

```python
# res=soup.find_all("a",{"class":"hwZGwa"})
# soup_st.find_all("a")
result = soup.find(id="JobCardGrid")
print(result.prettify())
```
