```python
import math
import matplotlib.pylab as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
t=np.arange(0.,1.,0.01)
```

```python
print(t)
```

```python
y1=np.sin(2*math.pi*t)
```

```python
plt.plot(t,y1)
```

```python
y2=np.sin(2*math.pi*4*t)
```

```python
plt.plot(t,y2)
```

```python
plt.plot(t,y2,'rx');
```

```python
plt.plot(t,y2,'r-.');
plt.grid(True)
```

```python
y3=np.cos(2*math.pi*4*t)
plt.plot(t,y2);
plt.plot(t,y3);
plt.show()
```

```python
plt.plot(t,y2,label='Sin')
plt.plot(t,y3,label='Cos')
plt.xlabel('time')
plt.ylabel('vlaue')
plt.title('Sin & Cos')
plt.legend()
plt.show()
```

```python
t2=np.arange(0.,1.,0.0001)
y2=np.sin(2*math.pi*4*t2)
y3=np.cos(2*math.pi*4*t2)
plt.plot(t2,y2,label='Sin')
plt.plot(t2,y3,label='Cos')
plt.xlabel('time')
plt.ylabel('vlaue')
plt.title('Sin & Cos')
plt.legend()
plt.show()
```
