```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.arange(0,3*np.pi,0.1)
y1=np.sin(x)

plt.plot(x,y1)
```

```python
y2=np.cos(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin and Cos')
plt.legend(['Sin','Cos'])
```

```python
plt.plot(x,y1,'rx',x,y2,'k:')
plt.legend(['Sin','Cos'])
```

```python
#plt.subplot(2,2,1)
x=np.arange(0,5,0.1)
y=np.sin(x)
y2=np.cos(x)
# print(xx)
plt.plot(x,y,'r--',x,y2,'k:')
```

```python
x=np.linspace(-1.4,1.4,30)
plt.subplot(2,2,1)
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.subplot(2,2,3)
plt.plot(x,x**3)
plt.subplot(2,2,4)
plt.plot(x,x**4)
plt.show()
```

```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2
```

```python
x=np.linspace(-2,2,500)
y=x**2
plt.plot(x,y,'r--')
plt.xlabel('$x$')
plt.ylabel('$y=x^2$')
plt.title('Sguare Function')
# save to file
plt.savefig('x.png')
```

```python
plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "sample point", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black", fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='black', ha="center", bbox=bbox_props)

plt.show()
```
