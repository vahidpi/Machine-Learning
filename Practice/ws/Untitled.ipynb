```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.arange(0,3*np.pi,0.1)
y1=np.sin(x)

plt.plot(x,y1)
```

```python
y2=np.cos(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin and Cos')
plt.legend(['Sin','Cos'])
```

```python
plt.plot(x,y1,'rx',x,y2,'k:')
plt.legend(['Sin','Cos'])
```

```python
xx=np.arange(0,5,0.1)
yy=np.sin(xx)
yy2=np.cos(xx)
# print(xx)
plt.plot(xx,yy,'r--',xx,yy2,'k:')
```
