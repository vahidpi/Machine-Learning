```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([[1,1],[1,2],[1,3]])
y=np.array([[1],[2],[3]])
theta=np.array([[1],[0]])
```

```python
plt.style.use('ggplot')

def plot_hyp(X,y,theta):
    # plot data
    plt.scatter(X[:,1],y,s=100,c='r',marker='X')
    
    x=np.linspace(0.8,3.2)
    x=x[:,None]
    x=np.hstack((np.ones_like(x),x))
    
    # plot hypothesis
    h=x @ theta
    plt.plot(x[:,1],h,c='k',lw=2)
    plt.show()
```

```python
plot_hyp(X,y,theta)
```

```python
def sse(X,y,theta):
    """ A vectorized implementation of sum of Squared Error
    """
    predictions = X @ theta
    errors = predictions -y
    return 0.5 * np.sum(errors**2) 
```

```python
def sse(X,y,theta):
    """ A vectorized implementation of sum of Squared Error
    """
    errors = X @ theta -y
    return 0.5 * (errors.T @ errors)[0,0]
```

```python
theta = np.array([[0],[0]])

plot_hyp(X,y,theta) ## print the line of prediction
loss=sse(X,y,theta) ## calculate error, J(θ)
print("J(%s) = %.2f" %(theta.ravel(),loss)) 
```

```python
theta = np.array([[0],[0.5]])

plot_hyp(X,y,theta) ## print the line of prediction
loss=sse(X,y,theta) ## calculate error, J(θ)
print("J(%s) = %.2f" %(theta.ravel(),loss)) 
```
